```python
#Convert TotalCharges column into float
dataset["SeniorCitizen"] = pd.to_numeric(dataset["SeniorCitizen"], errors="coerce")
```

```python
#Map all columns with yes and t=no to numerical values
variables = ["Partner", "Dependents", "PhoneService", "OnlineSecurity", "PaperlessBilling", "Churn", 
             "MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport",
            "StreamingTV", "StreamingMovies"]
#Mapping dictionary
mapping = {"Yes": 1, "No": 0}

#Convert variables column to numeric
for variable in variables:
    dataset[variable] = dataset[variable].map(mapping).fillna(2).astype(int)
```

```python
#Variables to encode
var = ["gender", "InternetService", "Contract", "PaymentMethod"]
encoder = LabelEncoder()
#Convert variables column to numeric
for variable in var:
    dataset[variable] = encoder.fit_transform(dataset[variable])
```
